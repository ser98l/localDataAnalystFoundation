```python
from sklearn.decomposition import PCA
pca= PCA(n_components=0.85)
pca.fit(df)
print (pca.explained_variance_ratio_)
print (pca.explained_variance_)
```

```python
#求相关系数矩阵
covX = np.around(np.corrcoef(df.T),decimals=4)#保留小数点后三位
covX
```

```python
featValue, featVec=  np.linalg.eig(covX.T)  #求解系数相关矩阵的特征值和特征向量(1×16)
featValue, featVec
```

```python
featValue = sorted(featValue)[::-1]#对特征值排序，特征值是对应主成分的方差
featValue
```

```python
np.around(featValue,decimals=4)
```

```python
# 同样的数据绘制散点图和折线图
plt.scatter(np.arange(1, df.shape[1] + 1), featValue)
plt.plot(np.arange(1, df.shape[1] + 1), featValue)

# 显示图的标题和xy轴的名字
# 最好使用英文，中文可能乱码
plt.title("Scree Plot")
plt.xlabel("Factors")
plt.ylabel("Eigenvalue")

plt.grid()  # 显示网格
plt.show()  # 显示图形
```

```python
#求特征值的贡献度
gx = featValue/np.sum(featValue)
gx
```

```python
lg = np.cumsum(gx)
lg
```

```python
#type(lg)
```

```python
gx_sum = 0.0
i = 0
while True:
   gx_sum += gx[i]
   i = i+1
   if gx_sum >=0.85:
      print(f'i = {i}\n'+f'gx_sum={gx_sum}')
      break
```

#取6个主成分使得累计贡献率达到85%

```python
#选出主成分
k=[i for i in range(len(lg)) if lg[i]<0.85]
#在 for 语句后面跟上一个 if 判断语句，用于过滤掉那些不满足条件的结果项。
k = list(k)
print(k)
```
