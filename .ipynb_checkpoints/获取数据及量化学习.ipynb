```python
#设置行列不忽略
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',20)
```

```python
#转换周期：日K转换为周K
#获取日K
#获取周K（当周的）：开盘价（当周的第一天）、收盘价(当周的最后一天)、最高价（当周）、最低价（当周）
df = get_price('000001.XSHE',count=100,end_date='2023-05-21', frequency='1d',panel=False)
df['weekday']=df.index.weekday
df
```

```python
1.355436e+09+1.150631e+09+7.917825e+08+8.761598e+08+7.307060e+08
```

```python
df_week = pd.DataFrame()
df_week['open'] =df['open'].resample('W').first()
df_week['close'] =df['close'].resample('W').last()
df_week['high'] =df['high'].resample('W').max()
df_week['low'] =df['low'].resample('W').min()
df_week['weekday']=df_week.index.weekday
df_week
```

```python
#汇总统计：统计月成交量、成交额（sum）
df_week['volume_sum'] = df['volume'].resample('W').sum()
df_week['money_sum'] = df['money'].resample('W').sum()
df_week
```

```python
# df_week2.drop(labels='date',axis=1,inplace=True)
from datetime import  *
from matplotlib.pylab import date2num
def date_to_num(dates):
    num_time=[]
    for i in dates:
        num_date = date2num(i)
        num_time.append(num_date)
    return num_time
    #Pythondate2num是matplotlib的一个模块,它将datetime类型的数据转换为matplotlib中所需要的float类型
```

```python
df_arr = df.values#转换为二维数组
df_arr[:,0]=date_to_num(df.iloc[:,0])#date_to_num函数更改日期格式
df_arr
```

```python
from mpl_finance import candlestick_ochl
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 8))

candlestick_ochl(ax, df_arr, width=0.4, colorup='r', colordown='g')
ax.xaxis_date()
ax.grid(True)

plt.title('K线图')
plt.xlabel('日期')
plt.ylabel('价格')
plt.show()
```
